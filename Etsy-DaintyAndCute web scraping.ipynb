```python
name=[]
price=[]
headers = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:102.0) Gecko/20100101 Firefox/102.0'}
for i in range(1,5):
    website='https://www.etsy.com/shop/DaintyAndCute?ref=items-pagination&sort_order=price_asc&page=' + str(i) + '#items'
    print(f"=================================loading {i}page==============================================")

    #request to website
    response=requests.get(url=website, headers=headers)
    
    code=response.status_code
    url=response.url
    print(url,code)

    #soup object
    soup=BeautifulSoup(response.content,'html.parser')
   
    #results
    results=soup.find_all('div',{'class':'v2-listing-card__info'})
  
    for a in results:
        try:
            name.append(a.find('h3',{'class':'wt-text-caption v2-listing-card__title wt-text-truncate'}).get_text().strip())  
        except:
            name.append('n/a') 
        try:
            price.append(a.find('span',{'class':'wt-screen-reader-only'}).get_text().strip()) 
        except:
            price.append('n/a')
```

```python
DaintyAndCute=pd.DataFrame({'Name':name,'Price':price})     
```

```python
name=[]
price=[]
Shopping_status=[]

for i in range(1,5):
    website='https://www.etsy.com/shop/DaintyAndCute?ref=items-pagination&sort_order=price_asc&page=' + str(i) + '#items'
    print(f"=================================loading {i}page==============================================")

    #request to website
    response=requests.get(url=website)
    
    code=response.status_code
    url=response.url
    print(url,code)

    #soup object
    soup=BeautifulSoup(response.content,'html.parser')
   
    #results
    results=soup.find_all('div',{'class':'v2-listing-card__info'})
  
    for a in results:
        try:
            name.append(a.find('h3',{'class':'wt-text-caption v2-listing-card__title wt-text-truncate'}).get_text().strip())  
        except:
            name.append('n/a') 
        try:
            price.append(a.find('span',{'class':'wt-screen-reader-only'}).get_text().strip()) 
        except:
            price.append('n/a')
        try:
            Shopping_status.append(a.find('div',{'class':'wt-text-brick wt-text-caption wt-pt-xs-1'}).get_text().strip())
        except:
            Shopping_status.append('n/a')
```

```python
d=pd.DataFrame({'Name':name,'Price':price,'Note':Shopping_status})     
```
