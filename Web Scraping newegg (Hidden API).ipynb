PUT EVERYTHING TOGETHER

```python
product_name=[]  
rating=[]  
rating_count=[]  
current_price=[]  
rebate=[]  
sale_end=[]
```

```python
#result_item 前面定义过的
for a in result_items:
    #product name
    try:
        product_name.append(a['Description']['ProductName'])
    except:
        product_name.append('n/a')
        
    #rating
    try:
        rating.append(a['Review']['Rating'])
    except:
        rating.append('n/a')
        
    #rating_count
    try:
        rating_count.append(a['Review']['HumanRating'])
    except:
        rating_count.append('n/a')
                
    #current_price
    try:
        current_price.append(a['FinalPrice'])
    except:
        current_price.append('n/a')    
        
    #rebate
    try:
        rebate.append(a['InstantRebateAmount'])
    except:
        rebate.append('n/a') 

    #sale_end
    try:
        sale_end.append(a['PromotionScheduleExpirationLocal'])
    except:
        sale_end.append('n/a') 
        
        
```

# change list to dataframe

```python
newegg_deal=pd.DataFrame({'Product Name':product_name,'Rating':rating,'Rating Count':rating_count,'Current Price':current_price,'Rebate':rebate,'Sale End':sale_end})             
```
